```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = "kribbeln kribbeln kribbeln zucken anstregend angenehm angenehm unangenehm unterstützend leistungsfördern irritierend irritierend irritierend verunsichernd lästig elektrisierend aufregend aufregend neuartig"

# Create the word cloud object
wordcloud = WordCloud(width = 800, height = 400, background_color='white').generate(text)

# Display the generated word cloud image
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
